```python
#sacado del ejercicio 3
def count(file, pattern):
    
    with open(file) as f: # con esto garantizo el cierre de f
    
        lines = f.readlines()
        count = 0

        for line in lines:
            if pattern in line:
                count += 1 
        print(f.closed)  # nunca cerrado
    print(f.closed)  # Garantizado cerrado
    
    return count

path = '/home/dsc/Data/shell/Finn.txt'

count(path, 'Huck')
```

```python
def process_a(lines):
    result = []
    
    for line in lines:
        if len(line) >1:
            result.append(last_word(line))
    return result

process_a(lines)
```

```python


def process_b(lines):
    result = []
    
    for line in lines:
        if len(line) >1:
            result.append(line[0]) # saca primera letra de cada linea
    return result

process_b(lines)
```

```python
def process(f, lines):
    result = []
    
    for line in lines:
        if len(line) >1:
            result.append(f(line))
    return result

process(len,lines) #(str.upper, lines) (word, lines)...etc
```
